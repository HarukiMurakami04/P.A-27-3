```python
# Estrutura de Decisões (if, else e elif)

#Escrever um programa que compara dois numeros

a = int (input("Digite um numero:"))
b = int (input("Digite um numero:"))

#Processamento / saida

if a == b:
  print (f"{a} é igual a {b}")

else:
  print (f"{a} é diferente de {b}")
```

```python
print(f"|---------Calculadora--------|")
print(f"|                            |")
print(f"|------------Menu------------|")
print(f"|----------------------------|")
print(f"|                            |")
print(f"|opção = 1 -soma +           |")
print(f"|opção = 2 -subt -           |")
print(f"|opção = 3 -mult ×           |")
print(f"|opção = 4 -divi ÷           |")


print(f"selecione os números:")
num1 = float(input(f"digite um numero:"))
num2 = float(input(f"digite um numero:"))

select = str(input(f"selecione o cálculo:"))

if select == "+":
  result1 = num1 + num2
  print(f"o resultado é {result1}")

elif select == "-":
  result2 = num1 - num2
  print(f"o resultado é {result2}")

elif select in ["X", "x", "×", "*"]:
  result3 = num1 * num2
  print(f"o resultado é {result3}")

elif select in [":", "÷", "/",]:
  result4 = num1 / num2
  print(f"o resultado é {result4}")

else:
  print (f"isso não é um cálculo")
```
